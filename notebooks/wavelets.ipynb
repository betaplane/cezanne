```python
import pandas as pd
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
%matplotlib nbagg
```

```python
with pd.HDFStore('../../data/CEAZAMet/station_data_new.h5') as S:
    p = S['pa_hpa']['PLV'].xs('prom', 1, 'aggr')['2016':].dropna().resample('h').first()
```

```python
cwt = sig.cwt(p.values.flatten(), sig.ricker, np.arange(24, 240))
```

```python
plt.figure()
plt.pcolormesh(cwt, vmin=-50, vmax=50)
```
